# Creates file handling methords

```python
def saveToFile(filename, abstract):
    if len(abstract) > 1:
        file = open("authors/" + filename, 'a', encoding="utf-8")
        try:
            file.writelines(abstract + "\n")
        except:
            file.write("\n")
            print("\n Write exception")
        file.close
        print('+')
        return True
    else:
        print('-')
        return False
```

```python
def createEmptyFile(filename):
    file = open("authors/" + filename, 'w', encoding="utf-8").close()
```

```python
def readFile(folder, filename):
    fileData = ""
    file = open(folder + "/" + filename, 'r', encoding='utf8')
    fileData = file.read().splitlines()
    file.close
    return fileData
```
