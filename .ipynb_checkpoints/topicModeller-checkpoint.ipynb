## Pre-processing of corpus
We use `nltk` for `stop_words` to grant us greater control. We also use the filters provided by `gensim` such as `strip_tags` and `strip_punctuation` to easily remove other redundent symbols.

```python
import nltk
import gensim
import ssl

# This try is fixing an issue with Mac's
try:
    _create_unverified_https_context = ssl._create_unverified_context
except AttributeError:
    pass
else:
    ssl._create_default_https_context = _create_unverified_https_context

nltk.download('stopwords')      # If this code block complains, comment out this line
from nltk.corpus import stopwords
from gensim.parsing.preprocessing import *

CUSTOM_FILTERS = [
    strip_tags,                 # Remove HTML tags and such
    strip_punctuation,          # Remove punctuation
    strip_multiple_whitespaces, # Remove unecessary space and such
    strip_numeric,              # Remove numbers
    lambda x: strip_short(x, 2) # Remove words shorter than 2 chars
]

stop_words = stopwords.words('english')
```

## Function to remove stop_words from corpus
Makes everything lower case and removes `stop_words`, if not in stop words, append to the clean corpus.

```python
def clean_doc(doc):
    return " ".join([word for word in doc.lower().split() if word not in stop_words])

def clean_corpus(corpus):
    preprocessed_corpus = [preprocess_string(doc.lower(), CUSTOM_FILTERS) for doc in corpus]
    clean_corpus = []
    for list in preprocessed_corpus:
        clean_corpus.append([word for word in list if word not in stop_words])
    return clean_corpus
```

## Cleans the corpus
First we remove all the words in `stop_words` from the corpus, and then we run the `corpus_without_stopwords` through the `CUSTOM_FILTERS` using the `preprocess_string` function, which also parses all strings into lists of words.

```python
corpus_without_stopwords = [clean_doc(doc) for doc in corpus]
```

```python
corpus_clean = [preprocess_string(doc, CUSTOM_FILTERS) for doc in corpus_without_stopwords]
```

## Convert our clean corpus to a dictionary

```python
from gensim import corpora

dictionary = corpora.Dictionary(corpus_clean)
```

```python
# for id, word in dictionary.iteritems():
#     print(id, word)
```

## Latent Dirichlet Allocation (LDA) Modelling

```python
# This is a list containing lists of number corresponding to a word
# Document vs word-frequency matrix
m1 = [dictionary.doc2bow(doc) for doc in corpus_clean] # document 2 bag of words

number_of_topics = 10
number_of_passes = 3
which_random_state = 0

LDA = gensim.models.ldamodel.LdaModel
ldamodel = LDA(
    m1, 
    num_topics = number_of_topics, 
    id2word = dictionary,
    passes = number_of_passes,
    random_state = which_random_state
)

m2 = ldamodel.print_topics() # This is M2 (Topics vs. Word-Weight Matrix)
m2_alternative = [ldamodel.show_topic(topicid = i) for i in range(0,number_of_topics)]
```

## Saving and loading model from disk

We save the model to disk, which allows us to store and load it.

When loaded we can use the model as a black box function and an object.

```python
from gensim.test.utils import datapath
from gensim.models.ldamodel import LdaModel

# Saving model to notebooks/test_model
temp_file = 'test-model/model'
ldamodel.save(temp_file)

ldamodel = LdaModel.load(temp_file)
```
